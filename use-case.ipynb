```python
import pandas as pd
import numpy as np 
import psi as psi
```

```python
# ==========================================
# EJEMPLO DE USO CON TUS DATOS DE SEGUROS
# ==========================================

# 1. Simulemos datos de Entrenamiento (2024 - Mix Normal)
data_2024 = {
    'Suma_Asegurada': np.random.normal(300000, 50000, 1000), # Autos promedio
    'Antiguedad_Vehiculo': np.random.randint(0, 15, 1000),
    'Zona': np.random.choice(['Norte', 'Centro', 'Sur'], 1000, p=[0.3, 0.4, 0.3]),
    'Marca': np.random.choice(['Nissan', 'Toyota', 'Ford'], 1000)
}
df_train = pd.DataFrame(data_2024)

# 2. Simulemos datos de Culiacán (2025 - Drift Extremo)
# NOTA: Aquí simulamos el "Pánico": Suma asegurada altísima y autos muy nuevos
data_2025 = {
    'Suma_Asegurada': np.random.normal(800000, 100000, 1000), # <--- DRIFT MASIVO (Autos caros)
    'Antiguedad_Vehiculo': np.random.randint(0, 3, 1000),     # <--- DRIFT (Solo autos nuevos)
    'Zona': ['Norte'] * 1000,                                 # <--- DRIFT (Solo una zona)
    'Marca': np.random.choice(['Nissan', 'Toyota', 'Ford'], 1000) # Sin drift
}
df_score = pd.DataFrame(data_2025)

# 3. Ejecutar el reporte
reporte = psi.get_psi_report(df_train, df_score)

print("--- Reporte de Estabilidad Poblacional (PSI) ---")
print(reporte)
```
